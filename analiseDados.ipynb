### Importar as bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carregar os dados das respostas do POSCOMP 2016
dados_respostas_2016 = pd.read_csv("exportacao_respostas_poscomp_2016.csv", sep=';', encoding='utf-8')
dados_respostas_2016 = dados_respostas_2016.set_index('NÚMERO DE INSCRIÇÃO')
```

```python
dados_gabaritos_2016 = pd.read_csv("Gabaritos_poscomp_2016.csv", sep=';', encoding='utf-8')
dados_gabaritos_2016 = dados_gabaritos_2016.set_index('COD CARGO')
```

```python
# Dados da homologação dos candidatos POSCOMP 2016
dados_homologados_2016 = pd.read_csv("homologados_dados_contatos_poscomp_2016.csv", sep=';', encoding='utf-8')
dados_homologados_2016 = dados_homologados_2016.set_index('ESTADO')
dados_homologados_2016
```

```python
#Notas do POSCOMP 2016
dados_notas_2016 = pd.read_csv("Notas_poscomp_2016.csv", sep=';', encoding='utf-8')
dados_notas_2016
```

## Salvar dados

```python
dados_gabaritos_2016.to_csv('dados/poscomp2016/dados_gabaritos_2016.csv')
```

```python
dados_homologados_2016.to_csv('dados/poscomp2016/dados_homologados_2016.csv')
dados_notas_2016.to_csv('dados/poscomp2016/dados_notas_2016.csv')
dados_respostas_2016.to_csv('dados/poscomp2016/dados_respostas_2016.csv')
```
