## Agrupamento das especializadades

```python
df_poscomp
```

```python
df = df_poscomp.query('presença != "n"')
```

```python
df = df.iloc[:,5:31]
```

```python
def calcular_percentagem_acertos(dados, coluna_acertos):
    dados["%acerto_" + coluna_acertos] = (
        dados[coluna_acertos] / dados[coluna_acertos].sum()
    ) * 100
```

```python
df
```

```python
# Exemplo de uso:
# Supondo que 'teste_dados' é o seu DataFrame
for i in df.columns[1:]:
    calcular_percentagem_acertos(df, i)
```

```python
def sumAxis(dados, columns_axis):
   return dados.groupby("especialidade")[columns_axis].sum().reset_index()
```

```python
colunas_para_processar = df.columns[df.columns.str.startswith("%acerto_")]
resultados_df = sumAxis(df, colunas_para_processar)
```

```python
resultados_df
```

```python
# Bases de dados filtradas com a porcentagens de acertos por eixos de cada estado



percentil_eixos = (
    df.groupby("especialidade")
    .agg(
        {
            "%acerto_matematica": "sum",
            "%acerto_fund_computacao": "sum",
            "%acerto_tec_computacao": "sum",
            # "%acerto_total": "sum",
        }
    )
    .reset_index()
)
```

```python
# Bases de dados filtradas com a porcentagens de acertos por eixos de cada sexo
percentil_sexo = (
    df.groupby("sexo")
    .agg(
        {
            "%acerto_matematica": "sum",
            "%acerto_fund_computacao": "sum",
            "%acerto_tec_computacao": "sum",
            # "%acerto_total": "sum",
        }
    )
    .reset_index()
)
```

```python
# Bases de dados filtradas com a porcentagens de acertos por eixos de cada região
percentil_estado = (
    df.groupby("estado")
    .agg(
        {
            "%acerto_matematica": "sum",
            "%acerto_fund_computacao": "sum",
            "%acerto_tec_computacao": "sum",
            # "%acerto_total": "sum",
        }
    )
    .reset_index()
)
```

```python
percentil_eixos.columns
```

```python
percentil_eixos = percentil_eixos.round(
    {"%acerto_matematica": 2, "%acerto_fund_computacao": 2, "%acerto_tec_computacao": 2}
)

percentil_estado = percentil_estado.round(
    {"%acerto_matematica": 2, "%acerto_fund_computacao": 2, "%acerto_tec_computacao": 2}
)
```
