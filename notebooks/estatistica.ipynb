```python
renda_estatisticas_por_uf = dataset_poscomp.groupby(['ESPECIALIDADE']).agg({'matematica': ['mean', 'median', 'max','min', 'std'],
                                                            'fund_computacao': ['mean', 'median', 'max','min', 'std'],
                                                            'tec_computacao': ['mean', 'median', 'max','min', 'std']})
renda_estatisticas_por_uf
```

```python
renda_estatisticas_por_uf = dataset_poscomp.groupby(['linhas_pesquisas']).agg({'matematica': ['mean', 'median', 'max','min', 'std'],
                                                            'fund_computacao': ['mean', 'median', 'max','min', 'std'],
                                                            'tec_computacao': ['mean', 'median', 'max','min', 'std']})
renda_estatisticas_por_uf
```

```python
renda_estatisticas_por_uf = dataset_poscomp.groupby(['area_concentration']).agg({'total': ['mean', 'median', 'max', 'std','sum']})
renda_estatisticas_por_uf = renda_estatisticas_por_uf.sort_values(by=('total', 'mean'), ascending=False)
renda_estatisticas_por_uf
```
