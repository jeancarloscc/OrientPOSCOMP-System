```python
X_dataset = dataset_poscomp.copy()
X_dataset = X_dataset.loc[:,['SEXO','ESTADO','IDADE','REGIAO','matematica', 'fund_computacao','tec_computacao', 'area_concentration']]
```

```python
# Aqui vamos pegar apenas os atributos MATEMATICA, FUNDAMENTOS_COMPUTACAO e TECNOLOGIA_COMPUTACAO
X = X_dataset.drop(['area_concentration'], axis=1)
y = X_dataset['area_concentration']
```

```python
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=10)

# enumerate the splits and summarize the distributions
for train_ix, test_ix in skf.split(X, y):
# select rows
    train_x, test_x = X.iloc[train_ix], X.loc[test_ix]
    train_y, test_y = y.iloc[train_ix], y.iloc[test_ix]

# summarize train and test composition
    train, train1 = len(train_y[train_y==0]), len(train_y[train_y==1])
    test, test1 = len(test_y[test_y==0]), len(test_y[test_y==1])
    print('>Train: 0=%d, 1=%d, Test: 0=%d, 1=%d' % (train, train1, test, test1))
```

```python
scaler = StandardScaler()

# Ajuste o scaler aos seus dados e, ao mesmo tempo, padronize-os
X = scaler.fit_transform(X)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42,stratify=y)
print("Shape of x_train : ",x_train.shape)
print("Shape of x_test  : ",x_test.shape)
print("Shape of y_train : ",y_train.shape)
print("Shape of y_test  : ",y_test.shape)
```

```python
def print_score(clf, x_train, y_train, x_test, y_test, train=True):
    if train:
        pred = clf.predict(x_train)
        clf_report = pd.DataFrame(classification_report(y_train, pred, output_dict=True))
        print("Train Result:\n==========================================================================")
        print(f"Accuracy Score: {accuracy_score(y_train, pred) * 100:.2f}%")
        print("__________________________________________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
        print("__________________________________________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_train, pred)}\n")
        
    elif train==False:
        pred = clf.predict(x_test)
        clf_report = pd.DataFrame(classification_report(y_test, pred, output_dict=True))
        print("Test Result:\n==========================================================================")        
        print(f"Accuracy Score: {accuracy_score(y_test, pred) * 100:.2f}%")
        print("__________________________________________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
        print("__________________________________________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_test, pred)}\n")
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parametros = {'criterion': ['gini', 'entropy', 'log_loss'],
              'splitter': ['best', 'random'],
              'min_samples_split': np.arange(8,20,2),
              'min_samples_leaf': np.arange(3,12),
             'max_depth':np.arange(1,10),
             'max_features': ['sqrt', 'log2']}
```

```python
tree_clf = DecisionTreeClassifier(random_state=10)
tree_clf = GridSearchCV(estimator=tree_clf, param_grid=parametros, scoring="accuracy", n_jobs=-1, verbose=1, cv=3)
tree_clf.fit(train_x, train_y)
melhores_parametros = tree_clf.best_params_
melhor_resultado = tree_clf.best_score_
print(melhores_parametros)
print(melhor_resultado)
print(tree_clf.best_estimator_)

tree_clf = DecisionTreeClassifier(**melhores_parametros)
tree_clf.fit(train_x, train_y)
print_score(tree_clf, train_x, train_y, test_x, test_y, train=True)
print_score(tree_clf, train_x, train_y, test_x, test_y, train=False)
```

```python
forest_clf = RandomForestClassifier(random_state=42)
forest_clf = GridSearchCV(estimator=forest_clf, param_grid=parametros, scoring="accuracy", n_jobs=-1, verbose=1, cv=3)
forest_clf.fit(train_x, train_y)
melhores_parametros = forest_clf.best_params_
melhor_resultado = forest_clf.best_score_
print(melhores_parametros)
print(melhor_resultado)
print(forest_clf.best_estimator_)

forest_clf = RandomForestClassifier(**melhores_parametros)
forest_clf.fit(train_x, train_y)
print_score(forest_clf, train_x, train_y, test_x, test_y, train=True)
print_score(forest_clf, train_x, train_y, test_x, test_y, train=False)
```

```python
parametros = [{'kernel':['poly'], 'degree':[0,1,2,3,4,5,6], 'C': [15,14,13,12,11,10,1,0.1,0.001]}]
svm_clf = SVC(random_state=42)
svm_clf = GridSearchCV(estimator=svm_clf, param_grid=parametros, scoring="accuracy", n_jobs=-1, verbose=1, cv=10)
svm_clf.fit(train_x, train_y)
melhores_parametros = svm_clf.best_params_
melhor_resultado = svm_clf.best_score_
print(melhores_parametros)
print(melhor_resultado)
print(svm_clf.best_estimator_)

svm_clf = SVC(**melhores_parametros)
svm_clf.fit(train_x, train_y)
print_score(svm_clf, train_x, train_y, test_x, test_y, train=True)
print_score(svm_clf, train_x, train_y, test_x, test_y, train=False)
```

```python
algorithms = [
    ('Árvore de Decisão (C4.5)', DecisionTreeClassifier(criterion='gini', max_features='sqrt',
                                                        max_depth=5, min_samples_leaf=9, min_samples_split=2, splitter='best')),
    ('Rede Neural', MLPClassifier(random_state=42, max_iter=500, learning_rate_init=0.01, activation='identity', hidden_layer_sizes=2, solver='adam')),
    ('Random Forest', RandomForestClassifier(random_state=42, n_estimators=100, criterion='gini', 
                                    min_samples_split=11, min_samples_leaf=3, max_features=3, max_depth=12, bootstrap=True)),
    ('SVM', SVC(random_state=42, kernel='linear', C=1)),
]

for name, classifier in algorithms:
    classifier.fit(train_x, train_y)
    predictions = classifier.predict(test_x)
    accuracy = accuracy_score(test_y, predictions)
    print(f'{name} - Acurácia: {accuracy * 100:.2f}%')
```
